# 3 处理缺失值
对于缺失值我们需要关注两个问题：

- 缺失值有多少
- 缺失值的随机的还是有它自己的模式

我们要去观察缺失值到底有多少并决定如何填充。

```python
# 查看缺失值
total_null = train_data.isnull().sum().sort_values(ascending=False)
percent = (train_data.isnull().sum() / train_data.isnull().count()).sort_values(ascending=False)
missing_info = pd.concat([total_null, percent], axis=1, keys=['Total', 'Null_Percent'])
```

```python
missing_info.head(20)
```

从上面的缺失值统计可以看到，缺失的特征有很多，因此我们要针对每个缺失的特征进行一一处理。

```python
# 定义两个函数
# 查看当前列的取值
def look_category(col):
    return train_data[col].value_counts()

# 填充空值位置
def fill_value(col, value):
    train_data.loc[train_data[col].isnull(), col] = value
```

# 5 分布检验
在现实问题中，很多时候数据都是右偏的状态，不服从正态分布，这样就会与我们所做的一系列假设进行冲突。为了能够让数据更加符合我们的假设，现实中都会对数据做一定的变换。

常见的做法是取对数，它的好处主要有：

- 1.研究的自变量数量级不一致时，取对数可消除这种数量级相差很大的情况。
- 2.取对数可以消除异方差。
- 3.取对数可以使非线性的变量关系转化为线性关系，更方便做参数估计。

```python
plt.subplot2grid((2,1),(0,0))
sns.distplot(train_data['SalePrice'])

# 对数变换
plt.subplot2grid((2,1),(1,0))
sns.distplot(np.log(train_data['SalePrice']))
```

从上面的两个图可以发现，在原始的数据上，saleprice是右偏分布，经过对数变换后，数据服从正态分布，均值大概在12左右。

## 转换数值属性
在这一部分我们主要完成三个任务：

- 对定序属性进行因子化
- 对非定序的变量进行one-hot
- 对连续值进行归一化

Note：这里有一个trick，就是对数据进行因子化和one-hot处理的时候，会出现训练集和测试集出现元素不一致的情况，也就是说有可能某个值在训练集中没出现，但是测试集中出现了，此时就会造成训练数据集和测试数据集中的特征不一致情况，因此在这里可以对训练数据集和测试数据集进行合并后处理非数值特征。然后再分开处理数值型特征。


```python
# 将训练数据分为特征和目标
train_data_features, train_data_targets = train_data.drop('SalePrice', axis=1), train_data['SalePrice']

# 训练样本个数
train_data_sample = train_data.shape[0]

# 合并所有特征
all_features = pd.concat([train_data_features, test_data])
```

```python
all_features.shape
```

```python
# 对需要one-hot编码的特征进行编码
onehot_features = ['MSSubClass', 'MSZoning', 'Street', 'Alley', 'Neighborhood', 'HouseStyle', 
                    'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation',
                    'CentralAir', 'SaleCondition'
                  ]
```

```python
# 对需要因子化的特征进行因子化
factorize_features = ['LandContour', 'Utilities', 'LotShape', 'LotConfig', 'LandSlope', 'Condition1',
                     'Condition2', 'OverallQual', 'OverallCond', 'ExterQual', 'ExterCond', 'BsmtQual',
                      'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC',
                      'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish',
                      'GarageQual', 'GarageCond', 'PavedDrive', 'Fence', 'YrSold',
                      'SaleType', 'BldgType', 'YearBuilt', 'YearRemodAdd']
```

```python
# 对连续特征处理
value_features = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF',
                 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath',
                 'BsmtHalfBath', 'FullBath', 'HalfBath', 'TotRmsAbvGrd', 'Fireplaces',
                 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF',  'OpenPorchSF', 'EnclosedPorch',
                 '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'BedroomAbvGr', 'KitchenAbvGr']
```

```python
# 检验特征数是否一样
if (len(onehot_features) + len(factorize_features) + len(value_features)) == (all_features.shape[1] - 1):
    print("Dimension Equal!")
```

```python
# 对onehot进行处理
for col in onehot_features:
    new_cols = pd.get_dummies(all_features[col], prefix=col)
    all_features = pd.concat([all_features, new_cols], axis=1)
    all_features.drop(col, axis=1, inplace=True)
```

```python
# 对因子化进行处理
for col in factorize_features:
    all_features[col] = pd.factorize(all_features[col])[0]
```

处理完非数值属性后，我们来观察与一下数值数据现在的状况

```python
all_features.loc[:, value_features].describe()
```

从上面的数据我们可以观察到，前面有一部分属性的取值非常大，这会对后面的带来一些误差，因为我们可以对其进行归一化，但对于小范围取值，例如1，2，3这种取值的变量不做处理。

```python
# 我们来看一下各个的最大取值
result = []
for each in value_features:
    result.append([each, np.max(all_features[each])])
                   
result = sorted(result, key=lambda x: x[1], reverse=True)
for r in result:
    print(r[0], r[1])
```

```python
# 从上面可以看到取值较大的数值变量，我们在这里把这些数值变量区分开
big_value_features = [r[0] for r in result if r[1] > 100]
```

```python
# 对这些特征进行log处理
for col in big_value_features:
    mean = all_features[col].mean()
    std = all_features[col].std()
    all_features[col] = (all_features[col] - mean) / std
```

```python
all_features.loc[:, value_features].describe()
```

# 建模
经过上面的数据处理，我们已经对数据进行了清洗和归一化操作，下面就可以进行建模来预测。
